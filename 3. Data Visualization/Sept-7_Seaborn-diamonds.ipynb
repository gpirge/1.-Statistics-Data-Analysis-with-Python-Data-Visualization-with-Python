# distplot

```python
sns.distplot(diamonds['carat']);
```

```python
sns.distplot(diamonds['carat'], bins = 15);
```

```python
sns.distplot(diamonds['carat'], kde = False, bins = 15, hist_kws={"linewidth": 2,
                            'edgecolor' : "k"});
```

```python
sns.distplot(diamonds['carat'], kde = False, bins = 25, hist_kws={"linewidth": 3,
                            'edgecolor' : "k"});
```

```python
sns.distplot(diamonds['carat'], hist = False, bins = 25, hist_kws={"linewidth": 3,
                            'edgecolor' : "k"});
```

```python
sns.distplot(diamonds['price'], kde = False, bins = 25, hist_kws={"linewidth": 1,
                            'edgecolor' : "k"});
```

```python
f, axes = plt.subplots(1, 2, figsize=(14, 7), sharex=True) # eksen limitlerini ayri ayri ayarlayamadim
sns.distplot(diamonds['price'], kde = False, bins = 25, hist_kws={"linewidth": 1,
                            'edgecolor' : "k"}, ax = axes[0])
#axes[0].set_xlim(0, 20000)
#axes[0].set_ylim(0, 16000)
sns.distplot(diamonds['carat'], kde = False, bins = 25, hist_kws={"linewidth": 3,
                            'edgecolor' : "k"}, ax = axes[1])
#axes[1].set_xlim(0, 5)
#axes[1].set_ylim(0, 16000);
```

# jointplot

```python
sns.jointplot(x='carat',y='price',data=diamonds,kind='reg');
```

```python
sns.jointplot(x='carat',y='price',data=diamonds,kind='hex');
```

# pairplot

```python
sns.pairplot(diamonds);
```

```python
sns.pairplot(diamonds,hue='cut',palette='coolwarm');
```

# barplot - a general plot that allows you to aggregate the categorical data based off some function

```python
#f, axes = plt.subplots(figsize=(14, 7))
sns.barplot(x='cut',y='price',data=diamonds); # default value is mean
```

```python
#f, axes = plt.subplots(figsize=(14, 7))
sns.barplot(x='cut',y='price',data=diamonds,estimator=np.mean);
```

```python
sns.barplot(x='cut',y='price',data=diamonds,estimator=np.max);
```

```python
sns.barplot(x='cut',y='price',data=diamonds,estimator=np.min);
```

```python
sns.barplot(x='cut',y='price',data=diamonds,estimator=np.median);
```

```python
sns.barplot(x='cut',y='price',data=diamonds,estimator=np.std);
```

```python
f, axes = plt.subplots(figsize=(14, 7)) #   of course y-axis contains the mean values
sns.barplot(x='cut',y='price',data=diamonds, hue = 'color'); 
```

# countplot - estimator is explicitly counting the number of occurrences

```python
sns.countplot(x='cut',data=diamonds);
```

```python
f, axes = plt.subplots(figsize=(14, 7))
sns.countplot(x='color',data=diamonds);
```

# boxplot - shows the distribution of categorical data

```python
f, axes = plt.subplots(figsize=(14, 7))
sns.boxplot(x="cut", y="price", data=diamonds,palette='rainbow');
```

```python
sns.boxplot(data=diamonds,palette='rainbow',orient='h'); #degerler cok orantisiz oldugu icin ...
```

```python
f, axes = plt.subplots(figsize=(14, 7))
sns.boxplot(x="carat", y="price", hue="color",data=diamonds, palette="coolwarm");
```

```python
sns.boxplot(x="cut", y="depth", data=diamonds,palette='rainbow');
```

# factorplot - most general form of a categorical plot. It can take in a kind parameter to adjust the plot type (renamed to catplot)

```python
sns.factorplot(x='carat',y='price',data=diamonds,kind='bar');
```
